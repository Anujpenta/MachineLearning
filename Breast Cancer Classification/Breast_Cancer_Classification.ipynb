```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

Dropping Id column.

```python
if "id" in train.columns:
    train = train.drop("id", axis=1)  
if "id" in test.columns:
    test = test.drop("id", axis=1)    
else:
    test_ids = np.arange(len(test))
```

```python
train.head()
```

Encoding target variable as M = 1 and B = 0

```python
label_encoder = LabelEncoder()
train['label'] = label_encoder.fit_transform(train['label'])
```

```python
# Split features and labels
X = train.drop('label', axis=1)
y = train['label']
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(test)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
y_val_pred = log_model.predict(X_train)
y_pred_log = log_model.predict(X_val)
print("Logistic Regression")
print("Accuracy:", accuracy_score(y_val, y_pred_log))
print(classification_report(y_val, y_pred_log))
```

```python
# SVM Model
svm_model = SVC()
svm_model.fit(X_train, y_train)

y_pred_svm = svm_model.predict(X_val)
print("SVM")
print("Accuracy:", accuracy_score(y_val, y_pred_svm))
print(classification_report(y_val, y_pred_svm))
```

```python
# XGBoost Model
xgb_model = XGBClassifier()

xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_val)

print("XGBoost Model")
print("Accuracy:", accuracy_score(y_val, y_pred))
```

```python
# For Random Forest Model
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
test_scaled = scaler.transform(test.drop(columns=['id'], errors='ignore'))
```

```python
rf = RandomForestClassifier()
rf.fit(X_train_scaled, y_train)
y_pred = rf.predict(X_val_scaled)

print("Validation Accuracy:", accuracy_score(y_val, y_pred))
```
