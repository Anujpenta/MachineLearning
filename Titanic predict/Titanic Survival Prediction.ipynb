Handling missing values

```python
# drop the cabin from the df

train = train.drop(columns="Cabin", axis = 1)
```

```python
test = test.drop(columns="Cabin", axis = 1)
```

```python
# replace missing values with mean age

train["Age"].fillna(train["Age"].mean(), inplace=True)
```

```python
test["Age"].fillna(test["Age"].mean(), inplace=True)
```

```python
# finding mode value of embarked coloumn

print(train["Embarked"].mode())
```

```python
print(train["Embarked"].mode()[0])
```

```python
print(test["Embarked"].mode())
```

```python
print(test["Embarked"].mode()[0])
```

```python
# replacing the embarked column with mode values

train["Embarked"].fillna(train["Embarked"].mode()[0], inplace=True)
```

```python
test["Embarked"].fillna(test["Embarked"].mode()[0], inplace=True)
```

```python
test["Fare"].fillna(test["Fare"].median(), inplace=True)
```

```python
# check no.of missing values in each column

train.isnull().sum()
```

```python
X = train.drop(columns=["PassengerId", "Name", "Ticket", "Survived"], axis = 1)
Y = train["Survived"]
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
# Model training
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
reg_train = model.predict(X_train)
```

```python
reg_train_acc = accuracy_score(reg_train, Y_train)
print(reg_train_acc)
```

```python
reg_test = model.predict(X_test)
```

```python
reg_test_acc = accuracy_score(reg_test, Y_test)
print(reg_test_acc)
```

```python
x = test.drop(columns=["PassengerId", "Name", "Ticket"], axis = 1)
```

```python
reg_actual = model.predict(x)
```

```python
submission = pd.DataFrame({
    "PassengerId": test["PassengerId"], "Survived": reg_actual
})

submission.head()
```
